```python
import requests

def check_robots_txt(url):
    robots_url = f"{url}/robots.txt"
    try:
        response = requests.get(robots_url)
        if response.status_code == 200:
            return response.text
        else:
            return f"Error: {response.status_code}"
    except requests.exceptions.RequestException as e:
        return f"Error: {e}"

# Example usage
url = "https://www.whitehouse.gov"
robots_txt = check_robots_txt(url)
print(robots_txt)
```

```python
# Initialize the main URL and maximum number of pages to scrape
main_url = 'https://www.whitehouse.gov/briefing-room/'
max_page = 1011
articles = []
previous_articles = None  # Store the articles from the previous page

# Iterate through each page until the maximum page number
for page in range(1, max_page + 1):
    print(f"Processing Page: {page}")

    # Construct the URL for each page
    page_url = main_url if page == 1 else f"{main_url}page/{page}/"
    
    # Attempt to request the webpage
    for attempt in range(2):
        try:
            time.sleep(random.randint(1, 3))  # Random delay to mimic human browsing
            response = requests.get(page_url, headers=get_random_user_agent())
            response.raise_for_status()
            break
        except requests.RequestException:
            if attempt == 1:
                print("Failed to retrieve the page after two attempts.")
                continue

    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')
    current_articles = []
    article_list = soup.find('div',class_='news-index').find('div',class_='container').find('div',class_='row justify-content-center articles-and-side-nav').find_all('article')
    
    # Check if any articles were found
    if not article_list:
        print("No articles found on this page.")
        break  # Stop if no articles are found
    
    # Extract details from each article
    for article in article_list:
        article_date = article.time.get('datetime', None)
        article_title = article.a.text.strip() if article.a else None
        article_url = article.a['href'] if article.a else None
        article_type = article.find('div',class_='news-item__meta shared-meta').a.text if article.find('div',class_='news-item__meta shared-meta').a.text else None
        current_articles.append((article_type, article_date, article_title, article_url))
    
    # Compare the current page's articles with the previous page's articles
    if current_articles == previous_articles:
        print("Duplicate page detected, stopping iteration.")
        break
    previous_articles = current_articles  # Update the previous articles list

    # Add the current page's articles to the main list
    articles.extend(current_articles)

# Save the results to a DataFrame and then to a CSV file
articles_df = pd.DataFrame(articles, columns=['article_type', 'article_date', 'article_title', 'article_url'])
articles_df.to_csv('whitehouse_articles.csv', index=False)
print("Scraping complete. Data saved to 'whitehouse_articles.csv'.")
```

```python
def request_attempt(url, attempt = 3):
    attempts = 0
    while attempts<attempt:
        try:
            time.sleep(random.randint(2,3))
            # yazar_page_request = requests.get(url,headers = get_random_user_agent(), proxies = {'http': proxy,'https': proxy}).text
            yazar_page_request = requests.get(url,headers = get_random_user_agent()).text

            return yazar_page_request
        except:
            attempts+=1
    return None
```

```python
def scrape_whitehouse_yazi(url):
    yazi_request = request_attempt(url)
    yazi_request_soup = BeautifulSoup(yazi_request, 'lxml')
    
    try:
        yazi_metin = yazi_request_soup.find('section',class_='body-content').find('div',class_='container').get_text(separator=' ', strip=True)
    except:
        yazi_metin = None
        
        
    return yazi_metin
```
